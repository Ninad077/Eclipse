# <font color = blue>**1. What is CNN?**</font>

CNN which is called as Convolutional Neural Network is majorly used for image classification. It is basically ANN but with some additional layers at the beginning. Some other domains where CNN is used are Scene labelling, object detection, face recognition, etc.

# <font color = blue>**2. Structure of CNN?**</font>

CNN is made up of 4 layers namely:


1.   Input layer
2.   Convolution layer
3.   Pooling layer
4.   Fully connected layer



# <font color = blue>**3. How does CNN work?**</font>

Image is made up of pixelated planes stacked on to each other. There are 3 planes Red, Green & Blue which are stacked together in a 3 dimensional space. So one plane has numerous amount of pixels depending upon the image dimensions. Each pixel cell carries a value from 0 to 255, 255 being the brightest of the colour plane and 0 being the darkest. One thing to remember here is that, Image being an unstructured data, is always passed as an array to the model.

The thing is, such pixelated Images at times carry pixels that are not important as well leading to a very slow building of the model. Hence in CNN, we have added two extra layers before the input layer namely Convolutional layer & Pooling layer. These additional layers extract only the essential pixels from the image.

# <font color = blue>**4. How does Convolution layer work?**</font>

A Convolution layer make use of matrix multiplication to make a convoluted image from a pixelated array. The convoluted image is smaller in dimension and contains only important pixels.

There's another concept called stride which one needs to know. So while doing matrix multiplication if the pixelated image is moving 1 pixel at a time then the stride= 1, if the pixelated image is moving 2 pixels at a time, then the stride = 2.

So Stride here is a hyperparameter. If one keeps stride=2 (more) then the model will take less time to train, however the extracted information will be less. On the contrary if stride=1 (less) then the model will take more time to train but the information extracted will be more. So, it is important to use optimum amount of Stride.

# <font color = blue>**5. How does Pooling layer work?**</font>

Pooling layer reduces the amount of information extracted in Convolution layer even further to keep only the most important pixels. So technically it reduces the feature size even further while retaining the most important information.
The two common functions used in Pooling layer are:


1.   Average pooling
2.   Max pooling

**In Average pooling we calculate the average/mean of each patch on the feature map.**

**In Max pooling we calculate the maximum value for each patch on the feature map.**

# <font color = blue>**6. What is Fully connected layer?**</font>

A Fully connected layer is the layer where the output from pooling layer is flattened up into a vector array. It is then passed on to neurons in the i/p layer, followed by transformations in dense layer and then the classes are predicted on the output layer. To put it briefly, this is where our ANN starts. So, one can say CNN has 2 extra layers Convolution & Pooling layer prior to ANN.

## <font color = Red>**1. Importing the libraries**</font>

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras. preprocessing.image import ImageDataGenerator
```

## <font color = Red>**2. Reading the image & defining it's dimensions**</font>

In the code block below, we are extracting dataset and defining batch size as 32 and image dimensions as 128x128.

```python
# Defining the dimensions of image
train_data_dir = 'dataset'
batch_size = 32
img_size= (128,128)
```

## <font color = Red>**3. Generating train & validation (test) data**</font>

We are using ImageDataGenerator function to rescale the image and dividing it into:

train = 80 %

validation (test) = 20% (validation_split = 0.2)

where the batch size and target image size is same as defined above.

'rescale=1./255' parameter is used to rescale the pixel values of the images. In this case, it is dividing all pixel values by 255. This operation scales the pixel values to the range [0, 1], which is typically done to normalize the input data. Normalization helps in improving the convergence of the training process.

```python
# Generating train data
datagen = ImageDataGenerator(rescale=1./255, validation_split= 0.2)

train_generator = datagen.flow_from_directory(
      train_data_dir,
      target_size = img_size,
      batch_size = batch_size,
      class_mode = 'categorical',
      subset = 'training'
    )
```

```python
# Generating validation data
validation_generator = datagen.flow_from_directory(
    train_data_dir,
    target_size = img_size,
    batch_size = batch_size,
    class_mode = 'categorical',
    subset = 'validation'
)
```

So overall we have got 1062 images in the training set and 264 images in the validation set. The classes here are the 3 image classes from the dataset namely Sea, Building & Forest.

## <font color = Red>**4. Creating the model**</font>

So we start by creating a Sequential model. In this model we add 32, 3x3 convolution layers. We have to define the input dimensions '128x128x3' as well.

It is then followed by adding 1, 2x2 pooling layer. I have created more such combinations of convolution & pooling layers to make sure the image is normalized even more.

It is then followed by flattening the image. Once flattened, the image will be passed as a 1D array to the i/p layer. Here I have then passed the image to 128 dense layers to generate an output with 3 dense layers.

There are 2 important things to note down here, the first being, since this is a classification model with more than 2 classes we are using activation function as 'softmax' or else if the classes were 2 we could have used 'sigmoid'.

Second being, we have introduced a Dropout layer between the dense and the output layer to avoid any unnecessary Overfitting of data.

```python
# Creating a Sequential model
model = models.Sequential()

# Adding  32 Convolution &  1 Pooling layer
model.add(layers.Conv2D(32,(3,3),activation = 'relu',input_shape = (128,128,3)))
model.add(layers.MaxPooling2D((2,2)))

# Adding 64 Convolution layers & 1 Pooling layer
model.add(layers.Conv2D(64,(3,3),activation = 'relu'))
model.add(layers.MaxPooling2D((2,2)))

# Adding 128 Convolution layers & 1 Pooling layer
model.add(layers.Conv2D(128,(3,3),activation = 'relu'))
model.add(layers.MaxPooling2D((2,2)))

# Flattening the layer to create a 1D array & passing it as an input.
model.add(layers.Flatten())
model.add(layers.Dense(128,activation = 'relu'))

# Adding a Dropout layer is used to avoid Overfitting & defining o/p layer.
model.add(layers.Dropout(0.5))
model.add(layers.Dense(3,activation = 'softmax'))
```

## <font color = Red>**5. Compiling the model**</font>

It is then followed by the traditional steps of ANN. Here, we compile the model on the optimizer 'adam', loss being 'categorical_crossentropy' (since we have a classification dataset) and metrics on which we compile is 'accuracy'.

```python
# Compiling the model
model.compile(optimizer = 'adam', loss='categorical_crossentropy',metrics = ['accuracy'])
```

## <font color = Red>**6. Training the model**</font>

We then train & fit the model on training data (train_generator) & validation data (test data). We define the epoch steps as the ratio of train generator samples (1062) divided by batch size (32) which would be 33 steps per epoch. We then define the total number of epochs required (100) followed by validation data & validation steps.

```python
# Training the model
history = model.fit(
    train_generator,
    steps_per_epoch = train_generator.samples // batch_size,
    epochs = 100,
    validation_data = validation_generator,
    validation_steps = validation_generator.samples // batch_size
)
```

## <font color = Red>**8. Testing the model**</font>

I have written a sample gradio code to test whether the model is working fine.

```python
import gradio as gr
import tensorflow as tf
import numpy as np

# Load the saved model
model = tf.keras.models.load_model('mymodel.h5')

# Define the labels
labels = ['Buildings', 'Forest', 'Sea']

# Define the image classification function
def classify_image(image):
    # Preprocess the image
    image = np.array(image)
    image = tf.image.resize(image, (128, 128))  # Resize the image to match the input size of the model
    image = tf.expand_dims(image, axis=0)  # Add batch dimension
    image = tf.keras.applications.resnet50.preprocess_input(image)

    # Predict the class
    predictions = model.predict(image).flatten()

    # Get the predicted class label
    predicted_class = labels[np.argmax(predictions)]

    return predicted_class

# Create the Gradio interface directly
interface = gr.Interface(classify_image, "image", "label", title="Image Classifier")

# Launch the interface
interface.launch()

```
