Now since this is a small data one can manually separate columns as well. However, we will try to automate the

```python
object_list=[]
int_list=[]
for i in data.columns:
    if data[i].dtype == "int64":
        int_list.append(i)
    else:
        object_list.append(i)
    
```

In the code above, I am creating a blank object list and integer list where I can append the segregated values.

integer (int64) I am appending the element in the blank integer list & for all the other cases I am appending it in the
object(string) list. 
This code would not work well if we have any other datatype other than object, because then for the esle condition, it will
append the datatype which is not an integer straight into the object list without checking whether it is an object in the

### <font color = red>18.3 fillna()</font>

'fillna()' is used to to fill the null values with the value specified inside it's argument.

Looking at the o/p above we have got null values for the column 'phd' and 'service'. Let us try to replace the null values 
the phd column with it's mean. The query would look like:

```python
data_new["phd"].fillna(data_new["phd"].mean())
```

Now adding an extra argument inplace = 'True' will substitute the null values with the specified value 
(i.e. mean in this example) not only in phd column, but also in the original dataset. 
So if we fetch the original datset back, the eliminated null values would be reflected in the table.

```python
data_new["phd"].fillna(data_new["phd"].mean(),inplace =True)
```

```python
data_new
```

Now after substituting null values with mean, let us check if we got any null values left.

```python
# Checking the null values for phd column now

data_new.isnull().sum()
```

As one can clearly see, the null values of phd column are eliminated (since, sum of 'phd' null values =0)

Similary, we can use fill the null values by 'mode' as well. 
To give an example:

```python
# Similary treating the null values of service column by it's mode

data_new["service"].fillna(data_new["service"].mode()[0],inplace=True)
```

### <font color = blue>20. Task: Write a code treating null values with mode if the column is categorical (dtype: object) and treating null values with mean if the column is numeric (dtype: int64 ordtype: int64 ) for the source "data_new" or in simple words <font color = red>'Create a Null Value function'</font></font>

Let us read the 'Salaries_new.xlsx' dataset on which we are going to perform Null value treatment.

```python
data_new= pd.read_excel("Salaries_new.xlsx")
```

```python
data_new
```

```python
# Checking the data info
data_new.info()
```

```python
# Checking the columns
data_new.columns
```

```python
def Null_Value_Treatment(data_new):
    for i in data_new.columns:
        if data_new[i].dtypes == "int64" or data_new[i].dtypes == "float64":
            data_new[i].fillna(data_new[i].mean(),inplace=True)
        else:
            data_new[i].fillna(data_new[i].mode()[0],inplace=True)
    return data_new.isnull().sum()
        
```

In the code above, we are first defining a 'Null value Treatment function'. Inside the function we are checking 
for every element (i) in data_new.columns (which in turnindicates we are checking for every column). 

In the next step we are defining 2 conditions. If the column (i.e. data_new[i]) has a datatype (dtype) integer 'int64' or
a float 'float64' then we are replacing the null values with the mean using 'fillna()' method. 

Else, we are replacing the null values with the mode using fillna() method.

In the end we are returning the total null value count of the dataset 'data_new' to check if the null values are treated
properly or not.

Let us call the function now.

```python
Null_Value_Treatment(data_new)
```

As evident above, we have successfully replaced the null values in each column using 'Null value treatment' function.

### <font color = blue>21. Task: Write a code treating null values with mode if the column is categorical (dtype: object) and treating null values with mean if the column is numeric (dtype: int64 ordtype: int64 ) for the source "df"</font>

The Task is similar to the one above. The only difference here is we are using a different dataset this time.

Let us read the dataset now.

```python
df=pd.read_csv("adult_new.csv")
```

```python
df
```

Now let us check the data type of all the columns.

```python
df.dtypes
```

```python
df.columns
```

Now let us define a Null Value Treatment function where we check the data type of all columns and replace the null
values in numeric columns (having data type float64 or int64) with mean and the null values in categorical columns by 
mode. The output of this function is the sum of total null values in the dataset (i.e. df.isnull().sum()).

```python
def Null_Value_Treatment(df):
    for i in df.columns:
        if df[i].dtypes == "int64" or df[i].dtypes == "float64":
            df[i].fillna(df[i].mean(),inplace=True)
        else:
            df[i].fillna(df[i].mode()[0],inplace=True)
    return df.isnull().sum()
```

```python
Null_Value_Treatment(df)
```
