## <font color = blue>9. Comment, Bold & Italics</font>

To comment in a code cell, one simply writes the comment prefixed by '#'. It might sound similar to a heading but
remember **# in heading is written in a 'markdown cell' whereas comment is written in a 'code cell'**.
In the code below **'# Adding 2 numbers'** is a comment that gives information about the underlying code.

```python
# Adding 2 numbers
1+1
```

To bold a word or a sentence in a markdown cell, simply use '**' before and after the word.

To give an example.

**This is a bold sentence**

Double click on this cell to understand how I have written the text above.

To make a word or sentence italics, use '***' before and after the word.

To give an example

***This is an italics sentence***

Double click on this cell to understand how I have written the text above.

## <font color = blue>10. Restarting a Kernel</font>

At times due to a server lag, the code might stop running or may take a lot of time to run. In such scenario,
one must restart the Kernel. To restart the kernel go to 'Kernel tab' and then you can either click on 'Restart' or
'Restart & Run All'.

I won't recommend 'Restart the Kernel' since all the output will be erased. So you have to run every cell again one by one.

Instead of doing this, I would strongly recommend you guys to select **'Restart & Run All'**, this will not only restart the kernel but run all the cells for you at once.
