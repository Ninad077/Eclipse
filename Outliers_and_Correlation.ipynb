## <font color = blue>3. Identifying Outliers using boxplot</font>

Let us plot a boxplot to understand the concept of outliers.

Outliers are the values that skew away from the original dataset. In simple words, the values that are a 
deviation from the original dataset.

Let us first start plotting boxplot for all the columns in data.

In the dataset, we have 3 columns namely 'TV', 'Newspaper' & 'Radio'. Let us plot the boxplot for each column one 
by one.

To plot a boxplot on TV column we use the syntax as follows:

```python
column = data["TV"]
plt.boxplot(column)
```

Let us understand the above o/p (boxplot) now.

So in the above boxplot, the upper and lower horizontal lines are called upper & lower whiskers.

The line inside the box is the median and the box itself indicates how many values lie close to the median.
For the above diagram, we've no outliers or in other words we do not have any values that are outside the boxplot on the 
extreme ends of the plot.

Let us check for another column 'Radio'.

```python
column = data["Radio"]
plt.boxplot(column)
```

Again we got a decent boxplot without any outliers. Now let us check the column 'Newspaper'.

```python
column = data["Newspaper"]
plt.boxplot(column)
```

As evident from the figure above, there are 2 co-ordinates that skew/deviate from the original data. These co-ordinates
are called as Outliers.

**Any values that lie beyond the upper and lower whiskers are called as Outliers.**

So now, if we want to eliminate the outliers in this case, we've to consider the data that lies below the threshold of
the upper whisker (which is 90).

```python
data_new = data[data["Newspaper"]< 90]
```

```python
data_new
```

Since, we are filtering the dataset to choose values less than 90, we should get a boxplot without outliers.
Let us check the boxplot now.

```python
plt.boxplot(data_new["Newspaper"])
```

This is one of the ways one can eliminate outliers from the data.

## <font color = blue>4. Understanding Correlation</font>

Correlation tells how different columns in the dataset are related to each other.

To define Correlation: One unit change in one coulmn will translate to how many units change in another column.
    
There are 3 types of Correlation:
    
    Positive correlation: Direct proportionality b/w 2 columns. 
                          If one increases another increases, if one decreases another decreases.
            
    Negative correlation: Inverse proportionality b/w 2 columns. 
                           If one increases other decreases, if one decreases other increases.
            
    No correlation      : No relation or dependancy b/w 2 columns. 
                          One's increase or decrease does not affect the other column.

### <font color = red>4.1 Positive Correlation & No Correlation</font>

To check for an example of positive correlation, let us see the following example.

Here, we are performing EDA using matplotlib. This time we are plotting a scatter plot.

Initially we define column_a & column_b as dataframe consisting of 'TV' & 'Sales' data.

It is then followed by plotting a scatter plot of column_a & column_b. 

The syntax would be: **plt.scatter(Co-ordinates_from_X_axis,Co-ordinates_from_Y_axis)**

It is then followed by defining the x,y labels and lastly the title of the plot.

```python
column_a = data["TV"]
column_b = data["Sales"]

plt.scatter(column_a,column_b)
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title("TV vs Sales: Positive Correlation")
```

Here as evident from the o/p above there exists are a strong linear correlation b/w TV and Sales. This indicates that
with increase in TV revenue, Sales is increasing as well. This is what we define as a Positive correlation.

Let us check the same for the column "Radio" & "Sales"

```python
column_c = data["Radio"]
column_d = data["Sales"]

plt.scatter(column_c,column_d)
plt.xlabel("Radio")
plt.ylabel("Sales")
plt.title("Radio vs Sales")
```

The data is all scattered and it clearly seems there is no correlation between 'Radio' & 'Sales'.

Let us check for the columns 'Newspaper' & 'Sales' now.

```python
column_e = data["Newspaper"]
column_f = data["Sales"]

plt.scatter(column_e,column_f)
plt.xlabel("Newspaper")
plt.ylabel("Sales")
plt.title("Newspaper vs Sales")
```

Similarly here as well, the data is all scattered and we cannot see any correlation between 'Newspaper' & 'Sales'.

### <font color = red>4.2 Heatmap</font>

One alternate EDA which is more frequently used to check correlation is called as 'heatmap'.

Heatmap is a matrix plot of correlations between columns within the data.

In a heatmap, The value of Correlation spans b/w -1 0 and 1: 

**-1 or closer values: Negative correlation**

**0 or closer values: No correlation**

**1 or closer values: Positive correlation**

We use seaborn as a library to visualize heatmap. The syntax used is : **sns.heatmap(dataset_name.corr(), annot= True)**.

.corr() is a method used to plot the correlation matrix, whereas annot = True is used to annote values (dispay the correlation values) on the matrix.



```python
import seaborn as sns
sns.heatmap(data.corr(), annot=True) # annot means annotation, it will assign values to the heatmap.
```

Let us interpret the heatmap above.

The strip on it's right side gives us the value of correlation.
Lighter shade (values closer to 1.0) gives higher correlation and Darker shade (values closer to 0.2) gives 
lower correlation.

The data in the diagonal will always have correlation 1 because it shows the correlation of the column with itself.
For e.g Sales (X-axis) = Sales (Y- axis), Newspaper (X-axis) = Newspaper (Y- axis)... and so on....

If we check the correlation of different columns wrt Sales we get,

    0.16 for Newspaper
    
    0.35 for Radio
    
    0.9 for TV (close to 1 hence a good positive correlation)
    
    
From the data above one can conclude that if one invests in marketing on 'TV' products the Sales will increase for sure.
